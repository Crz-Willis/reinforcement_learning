```python
import time
import numpy as np
import matplotlib.pyplot as plt
```

```python
l=[]
for i in range(10000):
    l.append(np.random.standard_normal())
```

```python
time_average=[]
time_incremental=[]
```

```python
start_time=time.time()
average=[0]
for i in range(1,10000):
    avg=average[i-1]+1.0/(i+1)*(l[i]-average[i-1])
    average.append(avg)
    t=time.time()
    time_incremental.append(t-start_time)
```

```python
start_time=time.time()
average2=[0]
for i in range(1,10000):
    avg=0
    for j in range(i):
        avg=avg+l[i]
    avg/=(i+1)
    average2.append(avg)
    t=time.time()
    time_average.append(t-start_time)
```

```python
x=[]
for i in range(9999):
    x.append(i)
plt.plot(x,time_average)
plt.plot(x,time_incremental)
```
